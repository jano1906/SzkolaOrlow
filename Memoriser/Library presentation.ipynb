```python
import Memoriser.DataGenerator.figures as dgf
import Memoriser.Grids.grids as gs
import Memoriser.Transforms.gaussian as gau
import Memoriser.memoriser as mm
```

# Library review

### Quick default datasets

```python
D=2
NO_DPOINTS = 1000
```

```python
centers, labels, rs = dgf.uniform_random_figures(D=D)
cdata = dgf.uniform(centers, labels, rs, D=D, n=NO_DPOINTS)
if D == 2:
    dgf.plot2D_data(cdata)
elif D == 3:
    dgf.plot3D_data(cdata)
```

```python
centers, labels, rs = dgf.normal_random_figures(D=D)
ndata = dgf.normal(centers, labels, rs, D=D, n=NO_DPOINTS, SIG=0.5)
if D == 2:
    dgf.plot2D_data(ndata)
elif D == 3:
    dgf.plot3D_data(ndata)
```

### Grids

```python
GSIZE = 1000
```

```python
cube = list()
gs.cube(cube, D=D, n = GSIZE)
if D == 2:
    dgf.plot2D_xc(cube)
elif D == 3:
    dgf.plot3D_xc(cube)
else:
    print(len(cube))
```

```python
ball = list()
scale = gs.ball(ball, D=D, n = GSIZE, r=0.5)
if D == 2:
    dgf.plot2D_xc(ball)
elif D == 3:
    dgf.plot3D_xc(ball)
else:
    print(len(ball))
```

```python
dgf.plot2D_xc(gs.scale(scale, ball))
```

### Transforms

```python
gau.D = D
gau.MU = 0.
gau.update()
gau_cube = [gau.T(x) for x in ball]
if D == 2:
    dgf.plot2D_xc(gau_cube)
elif D == 3:
    dgf.plot3D_xc(gau_cube)
```

```python
gau_cube_inv = [gau.Tinv(x) for x in gau_cube]
if D == 2:
    dgf.plot2D_xc(gau_cube_inv)
elif D == 3:
    dgf.plot3D_xc(gau_cube_inv)
```

```python
data_inv = [gau.Tinv(x[0]) for x in ndata]
if D == 2:
    dgf.plot2D_xc(data_inv)
elif D == 3:
    dgf.plot3D_xc(data_inv)
```
